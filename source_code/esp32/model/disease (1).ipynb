```python
zip_path = "disease_classification.zip"
extract_to = "disease_classification"
import os
from zipfile import ZipFile
os.makedirs(extract_to, exist_ok=True)
with ZipFile(zip_path, 'r') as z: z.extractall(extract_to)
print("Giải nén xong")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import os

# ======================================
# 1. Config
# ======================================
data_dir = "disease_classification"
train_dir = os.path.join(data_dir, "train")
val_dir = os.path.join(data_dir, "valid")

img_size = 224
batch_size = 32
num_classes = 4
num_epochs = 25
learning_rate = 1e-3

# ======================================
# 2. Tải dữ liệu
# ======================================
train_ds = keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=(img_size, img_size),
    batch_size=batch_size,
    label_mode="categorical"
)

val_ds = keras.utils.image_dataset_from_directory(
    val_dir,
    image_size=(img_size, img_size),
    batch_size=batch_size,
    label_mode="categorical"
)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.shuffle(1024).prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)

# ======================================
# 3. Augmentation (giống PyTorch)
# ======================================
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomFlip("vertical"),
    layers.RandomRotation(20/360),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.2),
    layers.RandomBrightness(0.2),
    layers.RandomSaturation(0.2),
])

# ======================================
# 4. EfficientNetV2-S pretrained
# ======================================
base_model = keras.applications.efficientnet_v2.EfficientNetV2S(
    include_top=False,
    input_shape=(img_size, img_size, 3),
    weights="imagenet"
)
base_model.trainable = True  # cho phép fine-tune

# Preprocessing chuẩn ImageNet
preprocess = keras.applications.efficientnet_v2.preprocess_input

inputs = layers.Input(shape=(img_size, img_size, 3))
x = data_augmentation(inputs)
x = preprocess(x)
x = base_model(x, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)

model = keras.Model(inputs, outputs)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# ======================================
# 5. Train
# ======================================
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=num_epochs
)

# ======================================
# 6. Vẽ đồ thị
# ======================================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss")
plt.legend(["Train","Val"])

plt.subplot(1,2,2)
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Accuracy")
plt.legend(["Train","Val"])
plt.show()

# ======================================
# 7. Xuất thành .tflite
# ======================================
print("\n🔄 Converting to TFLite...")

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # giảm kích thước model
tflite_model = converter.convert()

with open("disease_model.tflite", "wb") as f:
    f.write(tflite_model)

print("✅ Model saved as: disease_model.tflite")
```
